```python
lista_pets = ['Cachorro',
              'Pássaro',
              'Gato',
              'Coelho',
              'Peixe']
```

```python
# Função - Exibir Pets
def ver_pets():
  print(f'Os pets são: {lista_pets}')

def cadastrar_pets():
  novo_pet = input('Informe o pet: ')
  lista_pets.append(novo_pet)
```

```python
print('---PETS---')
print('Escolha a opção')
print('1 - Ver Pets')
print('2 - Cadastrar Novo Pet')
opcao = int(input('Informe a opção desejada: '))

if opcao == 1:
  ver_pets()
elif opcao == 2:
  cadastrar_pets()
  ver_pets()
else:
  print('Opção não encontrada')
```
